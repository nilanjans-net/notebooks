# Audience Interest Bar Chart using artist layer of Matplotlib

Survey on audience interest in different data science topics.

A survey was conducted to gauge an audience interest in different data science topics, namely:

1. Big Data (Spark / Hadoop)
2. Data Analysis / Statistics
3. Data Journalism
4. Data Visualization
5. Deep Learning
6. achine Learning

The participants had three options for each topic: Very Interested, Somewhat interested, and Not interested. 2,233 respondents completed the survey.

The survey results have been saved in a csv file and can be accessed through this link: https://cocl.us/datascience_survey_data.

If we examine the csv file, we will find that the first column represents the data science topics and the first row represents the choices for each topic.

We use the pandas read_csv method to read the csv file into a pandas dataframe:

In order to read the data into a dataframe we use the index_col parameter in order to load the first column as the index of the dataframe. 
Here is the documentation on the pandas read_csv method: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_csv.html

## Load packages

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline 
```

## Load the dataset

```python
import pandas as pd
df = pd.read_csv('https://cocl.us/datascience_survey_data', index_col='Unnamed: 0')
df
```

## Bar Chart using artist layer of Matplotlib

Now we use the artist layer of Matplotlib to draw the bar chart to visualize the percentage of the respondents' interest in the different data science topics surveyed.

To create this bar chart, we can follow the following steps:

1. Sort the dataframe in descending order of Very interested.
2. Convert the numbers into percentages of the total number of respondents. Recall that 2,233 respondents completed the survey. Round percentages to 2 decimal places.
3. As for the chart:
    1. we use a figure size of (20, 8),
    2. bar width of 0.8,
    3. we use color #5cb85c for the Very interested bars, color #5bc0de for the Somewhat interested bars, and color #d9534f for the Not interested bars,
    4. we use font size 14 for the bar labels, percentages, and legend,
    5. we use font size 16 for the title, and,
    6. we display the percentages above the bars, and remove the left, top, and right borders.

## Data preparation

```python
# Sort dataframe in descending order of "Very Interested"
df_new = df.sort_values('Very interested', ascending = False)

# Convert numbers into percentages
df_new = ((100 * df_new / 2233).round(2))
df_new
```

## Plot the chart

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline 

df_result = df_new / 100

ax = df_result.plot(kind = 'bar',
                figsize = (20,8),
                width = 0.8,
                color = ('#5cb85c', '#5bc0de', '#d9534f'),
                fontsize = 14)
ax.set_title("Percentage of Respondents' Interest in Data Science", fontsize = 16)
ax.legend(fontsize = 14)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.get_yaxis().set_visible(False)

# Format with percentage values    
for p in ax.patches:
    ax.annotate("{:.2%}".format(p.get_height()),
                xy=(p.get_x()+0.02, p.get_height()+0.01))
    
```
